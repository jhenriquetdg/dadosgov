```python
data_path = '../data'
```

```python
import os
```

```python
json_path = os.path.join(data_path,'universidade_federal_do_rio_grande_do_norte_-_ufrn.json')
```

```python
import json
```

```python
with open(json_path,'r') as file:
    json_data = json.load(file)
```

```python
from itertools import chain
```

```python
online_resources = list(chain(*[[l['link'] for l in d['recursos'] if 'formato' in l and l['formato'] == 'CSV'] for d in json_data['conjuntoDados']]))
```

```python
import re
```

```python
bolsistas = [r for r in online_resources if len(re.findall('bolsista',r))]
```

```python
import pandas as pd
```

```python
df_bolsistas = pd.concat(pd.read_csv(b,sep=';') for b in bolsistas)
```
